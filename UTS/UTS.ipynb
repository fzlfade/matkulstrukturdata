```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

def inorder(temp):
    if(not temp):
        return
    inorder(temp.left)
    print(temp.data, end=" ")
    inorder(temp.right)

def delDeep(root, d_node):
    q = []
    q.append(root)
    while(len(q)):
        temp = q.pop(0)
        if temp is d_node:
            temp = None
            return
        if temp.right:
            if temp.right is d_node:
                temp.right = None
                return
            else:
                q.append(temp.right)
        if temp.left:
            if temp.left is d_node:
                temp.left = None
                return
            else:
                q.append(temp.left)


def delet(root,key):
    if root == None:
        return None
    if root.left == None and root.right == None:
        if root.key == key:
            return None
        else:
            return root
    key_node = None
    q = []
    q.append(root)
    temp = None
    while(len(q)):
        temp = q.pop(0)
        if temp.data == key:
            key_node = temp
        if temp.left:
            q.append(temp.left)
        if temp.right:
            q.append(temp.right)
    if key_node:
        x = temp.data
        delDeep(root,temp)
        key_node.data = x
    return root


if __name__ == '__main__':
    root = Node(15)
    root.left = Node(3)
    root.left.left = Node(8)
    root.right = Node(7)
    root.right.left = Node(12)
    print("Tree sebelum proses hapus:")
    inorder(root)
    key = 12
    root = delet(root, key)
    print()
    print("Tree sesudah proses hapus:")
    inorder(root)
    print()
    print("Tree sesudah proses insert")
    root.right.right = Node(6)
    root.left.right = Node(5)
    inorder(root)
```

```python
class Node:
    def __init__(self, key, value):
        self.key = key
        self.value = value
        self.next = None

class separatechaining:
    def __init__(self, size):
        self.size = size
        self.table = [None] * size

    def hash_function(self, key):
        return hash(key) % self.size

    def insert(self, key, value):
        index = self.hash_function(key)
        if self.table[index] is None:
            self.table[index] = Node(key, value)
        else:
            current = self.table[index]
            while current.next:
                current = current.next
            current.next = Node(key, value)

    def search(self, key):
        index = self.hash_function(key)
        current = self.table[index]
        while current:
            if current.key == key:
                return current.value
            current = current.next
        return None

hash_table = separatechaining(10)
hash_table.insert("Paku", 5)
hash_table.insert("Palu", 10)
hash_table.insert("Gergaji", 15)
hash_table.insert("Skop", 20)
hash_table.insert("Pasah", 25)

print(hash_table.search("Skop"))
print(hash_table.search("Pasah"))
print(hash_table.search("Palu"))
```

```python
def bayar_utang(transactions):
    duit = {}
    for debtor, creditor, amount in transactions:
        duit[debtor] = duit.get(debtor, 0) - amount
        duit[creditor] = duit.get(creditor, 0) + amount

    utang = sorted([(person, -amount) for person, amount in duit.items() if amount < 0], key=lambda x: x[1])
    piutang = sorted([(person, amount) for person, amount in duit.items() if amount > 0], key=lambda x: x[1])

    settlements = []
    while utang and piutang:
        debtor, debt = utang.pop()
        creditor, credit = piutang.pop()

        payment = min(debt, credit)
        settlements.append((debtor, creditor, payment))

        debt -= payment
        credit -= payment

        if debt > 0:
            utang.append((debtor, debt))
        if credit > 0:
            piutang.append((creditor, credit))

    return settlements

transactions = [
    ('Ani', 'Budi', 5000),
    ('Ani', 'Doni', 10000),
    ('Budi', 'Cahyo', 10000),
    ('Cahyo', 'Ani', 1000),
    ('Cahyo', 'Eko', 10000),
    ('Doni', 'Cahyo', 5000),
    ('Eko', 'Budi', 5000),
    ('Eko', 'Doni', 500)
]

print(bayar_utang(transactions))
```
